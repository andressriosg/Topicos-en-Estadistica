```python
import numpy as np
import pandas as pd
```

```python
from google.colab import files # Para cargar archivos en google colab
uploaded = files.upload()
```

```python
import io
tabla = pd.read_excel(io.BytesIO(uploaded['CONSOLIDADO DATOS PRUEBA 55.xlsx']))
tabla.head(4)
```

```python
covariable = tabla['CONSUMO ACUM'] - tabla['CONSUMO ACUM'].mean(axis=0)
covariable = pd.DataFrame(covariable)
covariable.columns = ["covariable"]
covariable.head(5)
```

```python
tabla1 = pd.concat([tabla, covariable], axis=1)
tabla1.head(5)
```

```python
%pip install pingouin
from pingouin import ancova
```

```python
ancova(data=tabla1, dv='PESO', covar='covariable', between='REFERENCIA ALIMENTO')
```
