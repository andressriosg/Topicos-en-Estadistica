```python
import pandas as pd
import io
tabla = pd.read_csv(io.BytesIO(uploaded['BD-EEVV-Defuncionesfetales-2023.csv']), sep = ",", index_col=0, encoding='latin-1')
# tabla = pd.read_csv(io.BytesIO(uploaded['natalidad.csv']), sep = ";", index_col=0, encoding='latin-1')
tabla.head(4)
```

```python
tabla.columns
```

```python
df = tabla[["SEXO", "P_PMAN_IRIS", "PESO_NAC", "T_GES", "ASIS_MED"]]
df["P_PMAN_IRIS"] = df.P_PMAN_IRIS.astype(str)
df.head(4)
```

```python
df.loc[(df.P_PMAN_IRIS == "nan"), 'P_PMAN_IRIS'] = "6" # df[which(df$P_PMAN_IRIS == NA)] = "6"
df.head(4)
```

```python
df["ASIS_MED"] = df.ASIS_MED.astype(str)
df.head(4)
```

```python
df.loc[(df.ASIS_MED == "3"), 'ASIS_MED'] = "Ignorado"
df.loc[(df.ASIS_MED == "1"), 'ASIS_MED'] = "Si"
df.loc[(df.ASIS_MED == "2"), 'ASIS_MED'] = "No"
print(df)
```

```python
from sklearn.model_selection import train_test_split
x = df.drop("ASIS_MED", axis=1)
y = df["ASIS_MED"]
trainX, testX, trainY, testY = train_test_split(x, y, test_size = 0.2)
trainX
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaler = sc.fit(trainX)
trainX_scaled = scaler.transform(trainX)
testX_scaled = scaler.transform(testX)
testX.head(4)
```

```python
from sklearn.neural_network import MLPClassifier
mlp_clf = MLPClassifier(hidden_layer_sizes=(15,10,50),
                        max_iter = 300, activation = 'relu',
                        solver = 'lbfgs')
```

```python
mlp_clf.fit(trainX_scaled, trainY)
```

```python
from sklearn.metrics import accuracy_score
y_pred = mlp_clf.predict(testX)
print('Accuracy: {:.2f}'.format(accuracy_score(testY, y_pred)))
```
