# Cadenas

Las cadenas se pueden escribir en comillas simples o dobles.

```python
'huevos y pan'
"doesn't"
'"Si," le dijo.'
```

```python
s = 'Primera línea.\nSegunda línea.'
s
print(s)
```

En el intéprete, la salida de cadenas está encerrada en comillas y los caracteres especiales son escapados con barras invertidas. Aunque esto a veces luzca diferente de la entrada (las comillas que encierran pueden cambiar), las dos cadenas son equivalentes. Si no se quiere que los caracteres antepuestos por \ sean interpretados como caracteres especiales, se puede usar una cadena cruda agregando una r antes de la primera comilla.


```python
print('C:\\algun\\nombre')
print(r'C:\algun\nombre')
print(r'C:\\algun\\nombre')
```

Las cadenas de texto literales pueden contener múltiples líneas. Una forma es usar triple comillas: """...""" o
'''...'''. Los fin de línea son incluídos automáticamente, pero es posible prevenir esto agregando una \ al final de la
línea. Por ejemplo:

```python
print("""Uso: palabra 1
 palabra 2
 palabra 3""")
```

```python
print("""\\
Uso: palabra 1
 palabra 2
 palabra 3
 """)
```

# Listas

Python tiene varios tipos de datos compuestos, usados para agrupar otros valores. El más versátil es la lista, la cual puede
ser escrita como una lista de valores separados por coma (ítems) entre corchetes. Las listas pueden contener ítems de
diferentes tipos, pero usualmente los ítems son del mismo tipo.

```python
cuadrados = [1, 4, 9, 16, 25]
cuadrados
```

```python
print(cuadrados[0], cuadrados[-1])
```

```python
cuadrados + [36, 49, 64, 81, 100]
```

Las listas no son inmutables

```python
cubos = [1, 8, 27, 65, 125] # hay algo mal aquí, el cubo de 4 es 64, no 65!
cubos[3] = 64 # reemplazar el valor incorrecto
print(cubos)
```

```python
cubos.append(216) # agregar el cubo de 6
cubos.append(7 ** 3) # y el cubo de 7
print(cubos)
```

```python
letras = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
letras
```

```python
letras[2:5] = ['C', 'D', 'E'] # reemplazar algunos valores
letras
```

```python
letras[2:5] = [] # También se pueden borrar elementos
letras
```

```python
letras = ['a', 'b', 'c', 'd']
len(letras)
```

# Diccionarios

Los diccionarios guardan información en pares **clave : valor**.
No se accede por índice, sino por clave.

```python
alumno = {
    "nombre": "Juan",
    "edad": 18,
    "curso": "4° año"
}
alumno
```

```python
alumno["edad"]
```

```python
alumno["nombre"]
```

```python
len(alumno)
```

```python
alumno["edad"] = 19
alumno
```

```python
alumno["nota"] = 9 # Se pueden agragar nuevas claves a diccionarios
alumno
```

```python
del alumno["curso"] # También se pueden eliminar claves
alumno
```

```python
alumno["años"] = alumno["edad"] # Esta es una alternativa para cambiar de clave
alumno
```

```python
del alumno["edad"]
alumno
```

# Conjuntos

Los conjuntos **no tienen índices**. Es una colección de elementos
- sin orden
- sin elementos repetidos

```python
colores = {"rojo", "verde", "azul"}
colores
```

```python
colores1 = {"rojo", "verde", "azul", "verde"}
colores1
```

```python
colores1[0]
```

```python
for color in colores: # Para ver los elementos se puede usar un for
    print(color)
```

En Python existen distintos elementos para guardar y organizar información, y cada uno se usa según la necesidad.

Variables
Se usan cuando queremos guardar un solo valor (un número, un texto, un resultado).
Sirven para reutilizar datos y hacer el código más claro.

Listas
Se usan cuando necesitamos guardar varios valores ordenados y que puedan cambiar.
Son ideales cuando los datos pueden crecer, achicarse o modificarse.

Tuplas
Se usan cuando queremos guardar varios valores ordenados, pero que no deban modificarse.
Aportan seguridad: los datos no cambian accidentalmente.

Diccionarios
Se usan cuando necesitamos relacionar una clave con un valor.
Son útiles cuando los datos tienen significado (por ejemplo: nombre, edad, nota).

Conjuntos (sets)
Se usan cuando queremos guardar valores sin repetir y sin importar el orden.
Son útiles para comparar grupos y eliminar duplicados.

```python
lista = [1, 2, 3]
```

```python
tupla = tuple(lista)
tupla
```

```python
conjunto = set(lista)
conjunto
```

```python
lista1 = [1, 2, 2, 3]
set(lista1)   # {1, 2, 3}
```

```python
lista2 = [("nombre", "Ana"), ("edad", 20)]
diccionario = dict(lista2)
diccionario
```
