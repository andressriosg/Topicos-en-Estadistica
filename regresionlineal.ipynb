## Introducción

Se tienen dos variables de interés las cuáles se denotan por $X$ y $Y$. $X$ se denomina la variable *explicativa*, *regresora* o *independiente*, mientras que la variable $Y$ es la *variable respuesta* o *dependiente* de $X$. Implícitamente, bajo el enfoque de regresión (no únicamente lineal) se asume que hay dependencia entre $X$ y $Y$, adicionalmente que *Y* depende de cómo se comporta $X$.

El modelo de regresión más utilizado por su simplicidad y efectividad es el **modelo de regresión lineal simple**. La ecuación del modelo corresponde a:

$$Y = aX + b + \epsilon$$

donde $a$,$b$ son constantes y $\epsilon$ es una variable aleatoria con media cero ($E(\epsilon) = 0$) que describe la distancia entre el valor real de *y* y su valor estimado por $$\hat{y} = \hat{a}x+\hat{b}$$

Se tienen los siguientes *datos bivariados* $(x_1,y_1),...,(x_n,y_n)$, así como también los respectivos *residuales* (es decir, distancia de cada $y_i$ a $\hat{y}_i = \hat{a}x_i + \hat{b}$) dados por $\epsilon_1,...,\epsilon_n$.

En el modelo de regresión lineal simple se tienen los siguientes supuestos:

$(i)$ *Independencia*:  $\epsilon_1,...,\epsilon_n$ son independientes

$(ii)$ *Homocedasticidad*:  $\epsilon_1,...,\epsilon_n, \epsilon$ tienen la misma varianza dada por $\sigma^2$

$(iii)$ *Normalidad*:  $\epsilon_1,...,\epsilon_n, \epsilon \sim N(0,\sigma^2)$

Para profundizar más en el desarrollo teórico del modelo de regresión lineal simple, así como también estudiar la estimación de los parámetros del mismo se puede consultar el link de [regresión lineal simple](http://www4.ujaen.es/~dmontoro/Metodos/Tema%209.pdf).

Para el caso del modelo de regresión lineal múltiple, se tiene que $X$ es un vector aleatorio $k$-dimensional, es decir, se tienen $k$ variables explicativas. El modelo de regresión lineal múltiple se cumplen los mismos supuestos que el modelo de regresión lineal simple.

Es recomendable que $k < n$. En caso contario, se busca la *matriz inversa generalizada* de $X^{T}X$ (ver el [libro de modelos lineales](https://recursos.salonesvirtuales.com/assets/bloques/modeloslineales.pdf)).

## Marco histórico

Los cimientos del modelo de regresión lineal simple se basan fundamentalmente en experimentos de biología y de biometría. El primer autor interesado fue conocido como Adolphe Quetelet quien hizo sus aportes a partir de la estimación empírica de las medias y las desviaciones típicas de medidas estándar de las medidas antropométricas las cuales supuso que dependen de variables independientes tales como el sexo, la edad o incluso el nivel educativo.

En los estudios de Quetelet se relaciona la edad y la altura de las personas entre cero y 30 años por medio de la ecuación de una hipérbola. Su principal aporte fue haber considerado la dispersión en las medidas antropométricas, descubriendo además que estas satisfacen normalidad, distribución que modeló adecuadamente dicha variabilidad variabilidad (Ver [Seal, 1967](https://www.jstor.org/stable/2333849)).

Augusto Bravais contribuye al desarrollo a través de la astronomía al estudiar los errores en las medidas de las coordenadas de algunos cuerpos espaciales. Se le atribuye utilizar por primera vez el término *correlación* en 1846. En 1965 Pearson indicó que Bravais consideraba errores independientes pero no variables aleatorias correlacionadas por consiguiente no llegó a un concepto similar al de correlación que hoy en día tenemos.

Fue Galton quien estudió la variación de dos medias a través de sus estudios sobre la herencia y los estudios empíricos de sus leyes a través de métodos estadísticos. Galton describe algunas observaciones univariadas y bivariadas a través de distribuciones normales, en particular del parámetro $\rho$ de la distribución normal bivariada (y así teniendo un primer coeficiente de correlación). Luego de estudiar su teoría de la evolución Darwin propuso a Galton estudiar algún método con el fin de dar soporte a su teoría. Este método tenía que comparar las características físicas de los hijos con las de sus padres, esperando como resultado encontrar patrones hereditarios entre padres e hijos. Adicionalmente se requería establecer (o mejor, modelar) la variabilidad dentro de las características entre hijos y pares para poder sugerir las características de "peor" y "mejor adaptados" (para más detalles consultar [Benzecri, 1982](http://www.jehps.net/Decembre2008/Armatte.pdf)).

De esta manera Galton diseñó un experimento con semillas de guisantes para estudiar la distribución de sus pesos durante dos generaciones. Determinó que dicha distribución es aproximadamente normal, adicionalmente de concluir que:

(*1*) El peso medio de las semillas se explica linealmente a partir del peso de las semillas progenitoras con una pendiente menor que la unidad, es decir, si una semilla progenitora pesa $M+x$ entonces sus hijos pesarán $M + xr$ donde $r$ es una constante entre 0 y 1.

(*2*) La desviación estándar del peso de las semillas hijo es la misma para todos los grupos y es menor que la desviación estándar del peso de las semillas padres. Galtón acá encontró que las semillas filiales presentan *homocedasticidad*.

Para explicar estas propiedades Galton se basó en gráficos descriptivos. Galtón adicionalmente encontró la relación entre la varianza dentro (la varianza entre los individuos de estudio) y la varianza total (la variación de toda la población) a través del coeficiente de determinación. En 1869 Galton publica el libro [Hereditary Genius (Galton, 1869)](https://lesacreduprintemps19.files.wordpress.com/2012/11/galton-1869-genius-v3.pdf) donde estudia la influencia entre las características de padres e hijos, a través de un método llamado inicialmente *índice de correlación* el cual permite aproximar el parentesco entre un grupo de individuos expresarlo con un término único. Cuando no hay parentesco o relación familiar $r$ tiende a cero. Implícitamente Galton consideró la distribución conjunta de la variable $x$ la cuál es la medida común entre padres e hijos.

Por otra parte, Weldon calculó empíricamente coeficientes de correlación entre varias medidas físicas de órganos de camarones, es decir correlaciones entre medidas de los mismos individuos. También contribuyó en el desarrollo de la distribución muestral del coeficiente de correlación.

Más adelante Pearson continúa el trabajo de Galton y Weldon, quien en 1892 fue el primero en publicar señalando el significado de un coeficiente de correlación negativo. Galtón establecería el coeficiente de correlación como una razón cuyo valor va de -1 a 1 cuando un cambio en cualesquiera de los órganos implica un cambio igual en el otro y 0 cuando hay independencia entre las variables, como se puede leer en [Pearson, 1920](https://www.jstor.org/stable/2331722?seq=1). Pearson señaló “Galton no había aún alcanzado la idea de correlación negativa”. Adicionalmente Pearson dió la siguiente interpretación del coeficiente de correlación dado por las medidas antropométricas estudiadas por Galton: “la longitud del brazo se dice correlacionada con la de la pierna porque una persona de brazo largo usualmente tiene pierna larga y al contrario. Si la correlación es cercana, entonces una persona de brazos muy largos tendrá piernas muy largas; si es moderadamente próxima, entonces la longitud de la pierna sólo será larga, no muy larga; y si no hubiese ninguna correlación, entonces la longitud de la pierna sería, en promedio, mediocre” (ver [Estepa et al., 2012](https://core.ac.uk/download/pdf/20343739.pdf)).

Durante siglo XIX el trabajo de los estadísticos era mayormente descriptivo. La inferencia se desarrolla como consecuencia de la creación de la Escuela Biométrica del University College de Londres bajo la dirección de Pearson. Pearson demostró algunas propiedades del coeficiente de correlación y la regresión lineal simple, de hecho él la utilizó para la predicción de dos variables continuas. También generalizó la regresión estudiada por Galton a varias dimensiones, verificando que muchas variables biométricas siguen una distribución normal multivariada y por tanto se podría aplicar la regresión lineal múltiple.

Durante el mismo período de tiempo Yule introduce la notación y las ideas para la correlación parcial de dos variables dado una tercera variable independiente en el artículo [Yule, 1907](https://royalsocietypublishing.org/doi/10.1098/rspa.1907.0028).

El primer estudio documentado sobre el método de mínimos cuadrados es debido a Legendre quien usando optimización intenta encontrar la función que mejor se ajusta a datos bivariados correspondientes a la medición de meridianos utilizando como criterio el mínimo del error cuadrático, bajo el supuesto adicional de independencia de los datos. Legendre encontró que las ecuaciones funcionales se ajustaban adecuadamente esto debido a que los errores de medida hacía que los cálculos fuesen aproximados bajo el supuesto de normalidad de estos (ver más información detallada en [Sancho, 1980](https://dialnet.unirioja.es/descarga/articulo/4168462.pdf)).

En 1829 Gauss demostró el famoso *teorema de Gauss-Markov* (referencia [Reseña de Gauss, 2008](https://link.springer.com/referenceworkentry/10.1007%2F978-0-387-32833-1_159)) el cual muestra que los estimadores obtenidos con este método son insesgados y no se requiere de una distribución específica para lograr un ajuste de los datos. La aplicación del trabajo de Gauss, adicionalmente a los desarrollos de la actualidad, se evidenciaron posterioremente en 1899 en trabajos basados en el trabajo de Pearson y Yule, orientados a estudios sociales.

## El algoritmo de regresión lineal y su enfoque en *Machine learning*

La regresión lineal es un *algoritmo de aprendizaje de máquina supervisado* utilizado ampliamente en el análisis de datos. En su versión más sencilla, es decir, el *modelo de regresión lineal simple*, “dibuja una recta” donde se modela cómo depende la variable $Y$ de la variable $X$. Se asume que esta dependencia es de tipo lineal. Si los datos son discretos se puede utiliza un *modelo logístico* de acuerdo con este [enlace de regresión lineal](https://www.aprendemachinelearning.com/regresion-lineal-en-espanol-con-python/). En términos generales se tiene que un modelo lineal está dado por

$$Y = \textbf{X} \beta + \epsilon$$

donde $Y$ es la variable respuesta de dimensión $n \times 1$, $\textbf{X}$ es una matriz $n \times k$ cuyos coeficientes son conocidos, $\beta$ es un vector $k-$dimensional y $\epsilon$ es un vector $n-$dimensional el cual denota los errores (aleatorios). Se asume además que $\epsilon_1,...,\epsilon_n$ es una *muestra aleatoria* de $\epsilon \sim N(0, \sigma^2)$ (para más detalles teóricos consultar [Marín, 2012](http://halweb.uc3m.es/esp/Personal/personas/jmmarin/esp/Bayes/tema4bayes.pdf)).

Bajo enfoque de *Machine Learning* se asume que se tiene un *gran volumen de datos*. Al tener una suficiente cantidad de datos se pueden utilizar tantos *datos de entrenamiento* como *datos de prueba*. Con los datos de entrenamiento se construye el modelo y con los datos de prueba se determina si el modelo efectivamente es adecuado o no (por el ajuste de estos datos al mismo). Un método para establecer si un modelo se ajusta a los datos de prueba o no es el *coeficiente de determinación* dado por

$$R^2=\frac{(\hat{y}_i-\bar{y})^2}{(y_i-\bar{y})^2}$$

Hay diversos métodos para escoger los datos de entrenamiento y los datos de prueba: escogerlos aleatoriamente (usualmente 30% de los datos para prueba), hacer entrenamiento y prueba sobre los mismos datos, tomar varias muestras para entrenamiento y prueba de forma aleatoria, hacer *validación cruzada* o usar algún test de significancia estadística. El método de validación cruzada consiste en hacer varias divisiones de los datos. Los datos se dividen en datos de prueba y datos de entrenamiento. Al final se hace el entrenamiento a todos los datos destinados para ello de todas las divisiones y para la evaluación del modelo se toma el promedio de las medidas para la evaluación (por ejemplo, el *coeficiente de determinación*) calculadas sobre los datos de prueba de cada iteración o división realizada (para más detalles ver la [web de los métodos para escoger entrenamiento y prueba](https://machinelearningmastery.com/how-to-choose-the-right-test-options-when-evaluating-machine-learning-algorithms/) y [wikipedia](https://es.wikipedia.org/wiki/Validaci%C3%B3n_cruzada)).

La regresión lineal es un modelo de *Machine Learning Supervisado*, lo cual quiere decir que el modelo “aprende por sí mismo” (ver el [enlace de modelo lineal](https://www.aprendemachinelearning.com/regresion-lineal-en-espanol-con-python/)), es decir, a partir de los datos de entrenamiento se obtiene automáticamente la “recta” de regresión (se obtiene $\textbf{X} \beta$) con la cual se realizan predicciones usando el modelo: a partir de los datos de entrenamiento hay un aprendizaje para los valores de $\textbf{X}$ sin información.

Para estimar los valores de $\beta$ se debe minimizar la *función de costo* la cual corresponde al error cuadrático (para el modelo de regresión lineal), el cual es

$$e^2 = (y_i - X_i\hat{\beta})^2$$

Un método para minizar la función de costo es el *método del descenso del gradiente* el cual se basa en aproximar los gradientes $\partial (e^2/n)/\partial \beta$ a través de la recta $\beta - \alpha \partial (e_i^2/n)/\partial \beta$ calculada para algunos valores $e_i$ (ver la figura tomada del [link](https://www.iartificial.net/gradiente-descendiente-para-aprendizaje-automatico/))

```
from PIL import Image
import requests
from io import BytesIO

url = "https://www.iartificial.net/wp-content/uploads/2019/03/gradiente-descendiente-1024x641.webp"
response = requests.get(url)
img = Image.open(BytesIO(response.content))
basewidth = 400
wpercent = (basewidth / float(img.size[0]))
hsize = int((float(img.size[1]) * float(wpercent)))
img = img.resize((basewidth, hsize), Image.ANTIALIAS)
img
```

El valor de $\alpha$ es el *radio de aprendizaje* el cual se debe escoger ni muy grande porque puede no converger a algún mínimo (ver la siguiente figura tomada del [link](https://www.iartificial.net/gradiente-descendiente-para-aprendizaje-automatico/)), ni muy pequeño porque puede tardar demasiado en converger.

## Ejemplo con datos reales

Los datos de la tabla *natalidad.csv* corresponden a los datos de bebés recién nacidos en el año 2018 en la ciudad de Medellín. Los datos contienen información de todos los hospitales de dicha ciudad

```
from google.colab import files # Para cargar archivos en google colab
uploaded = files.upload()
```

```
import pandas as pd
import io
tabla = pd.read_csv(io.BytesIO(uploaded['natalidad.csv']), sep = ";")
tabla.head(3)
```

Tres variables sobre las cuáles se tiene interés corresponden a *peso del recién nacido* (en Kg), *edad en años de la madre* y *tiempo de gestación en semanas*. Se quiere determinar si el peso de un bebé recién nacido aumenta a medida que lo hacen tanto el tiempo de gestación y la edad de la madre durante el embarazo.

```
X, Y, Z = tabla["peso_nac"], tabla["t_ges"], tabla["edad_madre"]  # La variable regresora, la variable respuesta
print(X.head(3))
print(Y.head(3))
print(Z.head(3))
```

Notemos que hay datos atípicos (en realidad ilógicos porque un bebé recién nacido no puede pesar más de 6 kilos y su madre no puede tener más de 80 años) los cuáles se hace necesario eliminar

```
sin_atipicos = tabla[(tabla["peso_nac"] < 6000) & (tabla["t_ges"] < 80)]
print(sin_atipicos.head())
tabla1 = pd.concat([sin_atipicos["peso_nac"], sin_atipicos["t_ges"],  sin_atipicos["edad_madre"]], axis = 1)
print(tabla1.head())
```

Acá se quiere determinar en la medida en que el bebé pesa más al nacer ocurre cuando su madre tiene más años durante el embarazo. En el diagrama de dispersión se observa que si un bebé nace con mayor peso, su madre puede tener más edad.

```
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))
plt.scatter(tabla1["peso_nac"], tabla1["t_ges"])
plt.show()
```

Seleccionamos los datos de entrenamiento y de prueba. El criterio corresponde a seleccionar el 30% de los datos para prueba de manera aleatoria.  

```
from sklearn.model_selection import train_test_split # Acá vamos a seleccionar los datos de entrenamiento y de prueba
x_train, x_test, y_train, y_test, z_train, z_test= train_test_split(tabla1["peso_nac"], tabla1["t_ges"], tabla1["edad_madre"], test_size = 0.3)
 # Escogemos el 30% de los datos para prueba
print(x_train, y_train, z_train)
yz_train = pd.concat([y_train, z_train], axis=1)
print(yz_train)
yz_test = pd.concat([y_test, z_test], axis=1)
```

Ahora entrenamos el modelo con los datos de entrenamiento escogidos. Se utiliza el modelo de regresión lineal. En este caso dado por:

$$Peso_{nacido} = \beta_1 tiempo_{gestación}+ \beta_2 edad_{madre} + \epsilon$$

```
import numpy as np
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(yz_train, x_train)
```

Se pueden determinar los parámetros bajo el modelo de regresión lineal simple.

```
print("El intercepto del modelo es: ", modelo.intercept_)
print("Coeficiente:", list(zip(yz_train, modelo.coef_.flatten(), )))
```

También podemos determinar las medidas de ajuste sobre el parámetro.

```
print("Coeficiente de determinación R^2:", modelo.score(yz_test, x_test))
```

```
from sklearn.metrics import mean_squared_error
predicciones = modelo.predict(yz_test)
print(predicciones)
rmse = mean_squared_error(y_true  = x_test, y_pred  = predicciones, squared = True)
print("El cuadrado medio del error es: ", rmse)
```

*(1)*. El modelo de regresión es un modelo en el cual se asume que la variable respuesta $Y$ está en función de las variables $X_1, ..., X_n$ llamadas variables regresoras como $Y = X \beta + \epsilon$ donde $X = (\textbf{1}, X_1, ..., X_n)$ y $\epsilon$ es una variable aleatoria normalmente distribuida con media $0$ y varianza $\sigma^2$

*(2)*. Bajo el enfoque del aprendizaje de máquina se tienen dos tipos de datos: de entrenamiento y de prueba. Con los datos de entrenamiento se estiman los parámetros del modelo lineal y con los datos de prueba se verifica si el modelo tiene un ajuste adecuado o no (evaluación del modelo).

*(3)*. Para la estimación de parámetros se puede utilizar el método del descenso del gradiente, para el cual se determina numéricamente (usando métodos numéricos) la derivada parcial dada por $\partial e^2/n / \partial \beta$ a través de las aproximaciones $\beta - \alpha \partial (e_i^2/n)/\partial \beta$. Se debe escoger un radio de aprendizaje adecuado para garantizar la convergencia del método

*(4)*. A pesar de que en el ejemplo el modelo de regresión lineal escogido no es adecuado por tener un bajo coeficiente de determinación, sí indica que un tiempo de gestación mayor y una edad mayor de la madre, conduce a que un bebé recién nacido tenga mayor peso.
