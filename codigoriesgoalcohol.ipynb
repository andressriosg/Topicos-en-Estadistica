```python
dataf = pd.get_dummies(
    tabla_alcohol1,
    columns=['cerveza', 'herido', "preocupacion", "familia", "amigos"],
    prefix=['cerveza', 'herido',  "preocupacion", "familia", "amigos"]
)
dataf.head(4)
```

```python
dataf['cerveza_0.0'] = dataf['cerveza_0.0'].astype(int)
dataf['cerveza_1.0'] = dataf['cerveza_1.0'].astype(int)
dataf['cerveza_2.0'] = dataf['cerveza_2.0'].astype(int)
dataf['cerveza_3.0'] = dataf['cerveza_3.0'].astype(int)
dataf['cerveza_4.0'] = dataf['cerveza_4.0'].astype(int)
dataf['cerveza_5.0'] = dataf['cerveza_5.0'].astype(int)
dataf['herido_0.0'] = dataf['herido_0.0'].astype(int)
dataf['herido_1.0'] = dataf['herido_1.0'].astype(int)
dataf['herido_2.0'] = dataf['herido_2.0'].astype(int)
dataf["familia_0.0"] = dataf["familia_0.0"].astype(int)
dataf["familia_1.0"] = dataf["familia_1.0"].astype(int)
dataf["familia_2.0"] = dataf["familia_2.0"].astype(int)
dataf["amigos_0.0"] = dataf["amigos_0.0"].astype(int)
dataf["amigos_1.0"] = dataf["amigos_1.0"].astype(int)
dataf["amigos_2.0"] = dataf["amigos_2.0"].astype(int)
dataf["preocupacion_0.0"] = dataf["preocupacion_0.0"].astype(int)
dataf["preocupacion_1.0"] = dataf["preocupacion_1.0"].astype(int)
dataf["preocupacion_2.0"] = dataf["preocupacion_2.0"].astype(int)
dataf.head(4)
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(dataf, test_size=0.2, random_state=42)
```

```python
X_train = train_data.drop('recordar', axis=1) # Variables explicativas
y_train = train_data['recordar'] # Si la persona recuerda o no será la variable respuesta
X_test = test_data.drop('recordar', axis=1)
y_test = test_data['recordar']
```

```python
y_test.head(3)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
```

```python
y_pred = lda.predict(X_test) # Se hace prediccion sobre los datos de validacion para la variable explicativa
print("Predicciones:", y_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Calcular matriz
cm = confusion_matrix(y_test, y_pred)

# Crear figura
plt.figure(figsize=(6,5))

# Dibujar heatmap bonito
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    cbar=True,
    linewidths=1,
    linecolor='white'
)

# Etiquetas y título
plt.title('Matriz de Confusión', fontsize=14, fontweight='bold')
plt.xlabel('Predicción')
plt.ylabel('Valor Real')

plt.tight_layout()
plt.show()
```

De acuerdo con la matriz de confusión para valores de mayor riesgo, es decir, 3, 4 y 5, se encuentra que el algoritmo no hace predicciones adecuadas.

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

El 12 % de los datos de la categoría 3 se predicen en esa misma categoría, de manera análoga el 0% de la categoría 4 y el 3% de la categoría 5.

```python
from sklearn.ensemble import RandomForestClassifier

modelo_rf = RandomForestClassifier(
    n_estimators=300,
    class_weight='balanced',
    random_state=42
)

modelo_rf.fit(X_train, y_train)

y_pred_rf = modelo_rf.predict(X_test)

print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.linear_model import LogisticRegression

modelo_log = LogisticRegression(
    max_iter=1000,
    class_weight='balanced'
)

modelo_log.fit(X_train, y_train)

y_pred_log = modelo_log.predict(X_test)

print(classification_report(y_test, y_pred_log))
```

El modelo con la mejor capacidad de predicción de grupos vulnerables es el análisis discriminante lineal, con un acierto del 12% en la categoría 3 (versus el 2% de random forest y 6% de regresión logística), 0% en la categoría 4 (versus el 0% de random forest y 1% de regresión logística) y 3% en la categoría 5 (versus el 0% de random forest y 1% de regresión logística)
