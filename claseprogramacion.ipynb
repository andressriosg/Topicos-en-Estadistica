En Python existen distintos elementos para guardar y organizar información, y cada uno se usa según la necesidad.

Variables
Se usan cuando queremos guardar un solo valor (un número, un texto, un resultado).
Sirven para reutilizar datos y hacer el código más claro.

Listas
Se usan cuando necesitamos guardar varios valores ordenados y que puedan cambiar.
Son ideales cuando los datos pueden crecer, achicarse o modificarse.

Tuplas
Se usan cuando queremos guardar varios valores ordenados, pero que no deban modificarse.
Aportan seguridad: los datos no cambian accidentalmente.

Diccionarios
Se usan cuando necesitamos relacionar una clave con un valor.
Son útiles cuando los datos tienen significado (por ejemplo: nombre, edad, nota).

Conjuntos (sets)
Se usan cuando queremos guardar valores sin repetir y sin importar el orden.
Son útiles para comparar grupos y eliminar duplicados.

# Control de flujo en Python

```python
# Series de Fibonacci:
# la suma de dos elementos define el siguiente
a, b = 0, 1
while b < 100:
  print(b)
  a, b = b, a+b
```

* La primer línea contiene una asignación múltiple: las variables a y b toman en forma simultanea los nuevos valores 0
y 1. En la última linea esto se vuelve a usar, demostrando que las expresiones a la derecha son evaluadas antes de
que suceda cualquier asignación. Las expresiones a la derecha son evaluadas de izquierda a derecha.
* El bucle while se ejecuta mientras la condición (aquí: b < 100) sea verdadera. La prueba usada en el ejemplo es una comparación simple. Los operadores estándar de comparación se
escriben:
< (menor qué),
> (mayor qué),
== (igual a),
<= (menor o igual qué),
>= (mayor o igual qué)
!= (distinto a).
* El cuerpo del bucle está sangrado: la sangría es la forma que usa Python para agrupar declaraciones. En el intérprete
interactivo se debe teclear un tab o espacio(s) para cada línea sangrada. En la práctica, un buen editor, tiene la facilidad de agregar la sangría automáticamente. Al ingresar una declaración compuesta en forma interactiva, debe finalizar con una línea en blanco para indicar que está completa (ya que el analizador no puede adivinar cuando tecleaste la última línea).

```python
a, b = 0, 1
for _ in range(20):   # 6 repeticiones
    print(b)
    a, b = b, a + b
```

El tipo más conocido de sentencia es el *if*

```python
x = int(input("Ingresa un entero, por favor: "))
if x < 0:
  x = 0
  print('Negativo cambiado a cero')
elif x == 0:
  print('Cero')
elif x == 1:
    print('Uno')
else:
  print('Mayor que uno')
```

En este caso se clasifican números enteros. Puede haber cero o más bloques elif, y el bloque else es opcional. La palabra reservada 'elif' es una abreviación de 'else if', o en español "si no, entonces si".

En Python, break se usa para salir inmediatamente de un bucle (for o while) sin esperar a que termine normalmente.

```python
numeros = [3, 7, 12, 9, 5]
for n in numeros:
    if n == 12:
        print("¡Encontrado el número 12!")
        break
    print(n)
```

```python
numeros = [3, 7, 12, 9, 5]
for n in numeros:
    if n == 12:
        print("¡Encontrado el número 12!")
    print(n)
```

Se puede combinar for con if, elif y else

```python
for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
            print(n, 'es igual a', x, '*', n // x)
            break
    else:
        # Este else corresponde al for, no al if
        print(n, 'es un número primo')
```

try se usa para intentar ejecutar código que podría fallar, y capturar los errores sin que el programa se detenga.

```python
lista = [10, 20, 30]
try:
    index = int(input("Ingresa un índice: "))
    print("El valor es:", lista[index])
except IndexError:
    print("Índice fuera de rango")
except ValueError:
    print("Debes ingresar un número entero")
```

La palabra except se usa para capturar errores específicos que ocurren dentro del bloque try.

Cuando Python encuentra un error dentro de try, busca un except que coincida con el tipo de error.

Si lo encuentra, ejecuta el código dentro de ese except y el programa sigue funcionando, en lugar de detenerse.

```python
entrada = input("Ingresa un número: ")
try:
    numero = int(entrada)
    print("El número ingresado es:", numero)
except ValueError:
    print("Eso no es un número válido")
```
