```python
import pandas as pd
import io
serie_covid1 = io.BytesIO(uploaded['serie_completa (1).csv'])
serie_covid = pd.read_csv(serie_covid1, sep = ",", encoding="latin-1")

import matplotlib.pyplot as plt
serie_covid['Fecha'] = pd.to_datetime(serie_covid['Fecha'])
serie_covid = serie_covid.sort_values('Fecha').reset_index(drop=True) # Ordenar por fecha por si acaso
plt.figure(figsize=(10,4))
plt.plot(serie_covid['Fecha'], serie_covid['Frecuencia'])
plt.title('Serie COVID - Frecuencia diaria')
plt.xlabel('Fecha')
plt.ylabel('Frecuencia')
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
serie = serie_covid['Frecuencia']
plt.figure(figsize=(10,4))
plot_acf(serie.dropna(), lags=40, alpha=0.05)
plt.title("Funcin de Autocorrelacin (ACF)")
plt.xlabel("Rezagos")
plt.ylabel("Autocorrelacin")
plt.tight_layout()
plt.show()

# --- Funcin de autocorrelacin parcial (PACF) ---
plt.figure(figsize=(10,4))
plot_pacf(serie.dropna(), lags=40, alpha=0.05, method='ywm')
plt.title("Funcin de Autocorrelacin Parcial (PACF)")
plt.xlabel("Rezagos")
plt.ylabel("Autocorrelacin Parcial")
plt.tight_layout()
plt.show()
```

En la grfica de la funcin de autocorrelacin se observa que el nmero de casos de COVID-19 del da presente depende de cuntos casos se reportaron hace 40 das o ms. Descriptivamente se hace el diagrama de la autocorrelacin parcial y se encuentra que incluso 39 das antes se tiene que hay una dependencia neta de la serie hacia el tiempo presente. Esto quiere decir que se puede tratar de una serie estacional porque la autocorrelacin no considera el efecto de la serie sobre otros das distintos.

```python
from scipy.signal import periodogram, find_peaks
import pandas as pd
import numpy as np
import io

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import periodogram, find_peaks

x = serie_covid['Frecuencia'].values
fs = 1  # Frecuencia de muestreo: 1 observacin por da

# === Calcular el periodograma ===
f, Pxx = periodogram(x, fs=fs)

# === Detectar todos los picos significativos ===
# Prominence controla qu tan marcados son los picos (mejor que height)
peaks, properties = find_peaks(Pxx, prominence=np.max(Pxx)*0.005)  # 0.5% del mximo

# === Graficar periodograma ===
plt.figure(figsize=(12,5))
plt.plot(f, Pxx, color='steelblue', label='Periodograma')
plt.scatter(f[peaks], Pxx[peaks], color='red', s=40, label='Picos detectados')

# === Dibujar lneas verticales y etiquetas de periodo ===
for fp, power in zip(f[peaks], Pxx[peaks]):
    if fp > 0:
        periodo = 1 / fp
        plt.axvline(x=fp, color='gray', linestyle='--', alpha=0.4)
        plt.text(fp, power * 1.1, f'{periodo:.1f} das', rotation=90,
                 va='bottom', ha='center', fontsize=8, color='darkred')

plt.title('Periodograma con todos los picos detectados')
plt.xlabel('Frecuencia (ciclos por da)')
plt.ylabel('Densidad espectral de potencia')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

En el periodograma se observa que se trata de una serie estacional porque se observa que hay picos de frecuencia en este diagrama.

```python
periodos = 1 / f[peaks]
print("Frecuencias dominantes detectadas:")
for fp, per in zip(f[peaks], periodos):
    print(f"Frecuencia = {fp:.4f} ciclos/da    Perodo  {per:.2f} das")
```

Se toma un periodo de 55 das, lo que quiere decir que alrededor de cada dos meses una persona se reinfectar por COVID-19

```python
from statsmodels.tsa.seasonal import STL
periodo = 55 # <- periodo solicitado
seasonal = periodo  # en statsmodels.STL, seasonal debe ser odd; si es par, STL ajusta internamente
# Puedes ajustar robust=True para mayor resistencia a outliers
robust = True

# ========== APLICAR STL ==========
stl = STL(serie_covid['Frecuencia'], period=periodo, robust=robust)
res = stl.fit()

# ========== GRAFICAR RESULTADOS ==========
fig, axes = plt.subplots(4, 1, figsize=(12, 9), sharex=True)
axes[0].plot(serie_covid.index, serie_covid['Frecuencia'], label='Observada')
axes[0].set_ylabel('Observada')
axes[0].legend()

axes[1].plot(serie_covid.index, res.trend, color='orange', label='Tendencia')
axes[1].set_ylabel('Tendencia')
axes[1].legend()

axes[2].plot(serie_covid.index, res.seasonal, color='green', label='Estacional (periodo=55)')
axes[2].set_ylabel('Estacional')
axes[2].legend()

axes[3].plot(serie_covid.index, res.resid, color='red', label='Residuo')
axes[3].set_ylabel('Residuo')
axes[3].legend()

plt.suptitle(f"Descomposicin STL (periodo = {periodo}, robust={robust})", fontsize=14)
plt.xlabel('Fecha')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

Se observa que la tendencia de la serie es mejor modelarla considerando un periodo de 55 das, ya que con periodos ms prolongados, la tendencia de la serie no logra captar los picos de la pandemia. Al realizar la grfica de los residuales de la serie en la descomposicin se observa que se trata de un modelo heterocedstico.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.stats.diagnostic import acorr_ljungbox, het_arch

serie = serie_covid["Frecuencia"]
lb_test = acorr_ljungbox(serie, lags=[5, 10, 20, 55], return_df=True)
print("=== Prueba de LjungBox ===")
print(lb_test)
# H0: no hay autocorrelacin hasta ese nmero de rezagos
# Si p < 0.05  hay dependencia temporal significativa

# --- 3. Prueba de estacionariedad: DickeyFuller ---
adf_result = adfuller(serie)
print("\n=== Prueba ADF (DickeyFuller) ===")
print(f'Estadstico ADF: {adf_result[0]:.4f}')
print(f'p-valor: {adf_result[1]:.4f}')
# H0: la serie tiene raz unitaria (no estacionaria)
# Si p < 0.05  la serie es estacionaria

# --- 4. Prueba de heterocedasticidad: ARCHLM ---
arch_test = het_arch(serie)
print("\n=== Prueba de Heterocedasticidad (ARCHLM) ===")
print(f'Estadstico LM: {arch_test[0]:.4f}')
print(f'p-valor: {arch_test[1]:.4f}')
# H0: no hay heterocedasticidad (varianza constante)
# Si p < 0.05  la serie presenta heterocedasticidad
```

Cuando se tiene grandes volmenes de datos siempre se rechaza la hiptesis. De esta forma, se recomienda tomar varias muestras de los datos y a cada una se le realizar las pruebas de hiptesis.

```python
window_size = 100  # tamao de la ventana
n = len(serie)

# --- Resultados guardados ---
resultados = []

# --- Rolling window ---
for start in range(0, n - window_size + 1):
    end = start + window_size
    muestra = serie[start:end]

    # --- 1. Prueba de autocorrelacin: LjungBox ---
    lb_test = acorr_ljungbox(muestra, lags=[5, 10, 20, 55], return_df=True)
    lb_pvalues = lb_test['lb_pvalue'].values

    # --- 2. Prueba de estacionariedad: DickeyFuller ---
    adf_result = adfuller(muestra)
    adf_stat, adf_pvalue = adf_result[0], adf_result[1]

    # --- 3. Prueba de heterocedasticidad: ARCHLM ---
    arch_stat, arch_pvalue, _, _ = het_arch(muestra)

    # --- Guardar resultados ---
    resultados.append({
        'Inicio': start,
        'Fin': end,
        'ADF_pvalue': adf_pvalue,
        'ARCH_pvalue': arch_pvalue,
        'LjungBox_p5': lb_pvalues[0],
        'LjungBox_p10': lb_pvalues[1] if len(lb_pvalues) > 1 else np.nan,
        'LjungBox_p20': lb_pvalues[2] if len(lb_pvalues) > 2 else np.nan,
        'LjungBox_p40': lb_pvalues[3] if len(lb_pvalues) > 3 else np.nan
    })

# --- Convertir a DataFrame ---
resultados_df = pd.DataFrame(resultados)

# --- Mostrar resumen ---
print(resultados_df.head())
```

Las muestras construidas corresponden a 100 pasos continuos de la serie original. Deben ser continuos porque se trata de una serie de tiempo, por lo tanto, las variables aleatorias estn ordenadas.

```python
columnas_pval = ['ADF_pvalue', 'ARCH_pvalue',
                 'LjungBox_p5', 'LjungBox_p10', 'LjungBox_p20', 'LjungBox_p40']

# --- Crear el diagrama de cajas ---
plt.figure(figsize=(10,6))
plt.boxplot(resultados_df[columnas_pval].dropna().values, labels=columnas_pval, patch_artist=True)

# --- Personalizacin del grfico ---
plt.title('Diagramas de cajas y bigotes de los valores p por prueba', fontsize=14)
plt.ylabel('p-valor', fontsize=12)
plt.xticks(rotation=30)
plt.axhline(0.05, color='red', linestyle='--', linewidth=1.5, label='Nivel de significancia 0.05')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

Observando los diagramas de cajas de los valores p de las muestras de la serie de tiempo, se observa que no se rechaza con una confianza del 95% que la serie tenga races unitarias, es decir, el proceso no es estacionario. De manera anloga, se rechaza con una confianza del 95% que el proceso es homocedstico. No se rechaza con una confianza que la serie dependa de s misma tomando 5 das antes, 10 das antes, 20 das antes y 55 das antes.

```python
!pip install statsmodels --upgrade
from statsmodels.tsa.statespace.sarimax import SARIMAX
# Parmetros del modelo: (p,d,q)  (P,D,Q,s)
# En este ejemplo usamos el modelo de lneas areas (0,1,1)x(0,1,1,55)
# donde s=7 sugiere un patrn semanal
modelo = SARIMAX(serie_covid['Frecuencia'],
                 order=(5,1,1),            # ARIMA no estacional
                 seasonal_order=(1,1,1,55), # componente estacional multiplicativa
                 enforce_stationarity=False,
                 enforce_invertibility=False)

resultado = modelo.fit(disp=False)

# ============================================
# 4. Mostrar resumen de parmetros estimados
# ============================================
print("\n=== Estimadores de los parmetros SARIMA ===")
print(resultado.summary())
```

Se sugiere que los ordenes de la parte regular y estacional no sumen ms de 2 para evitar sobre ajuste.

```python
from sklearn.metrics import r2_score
fitted_values = resultado.fittedvalues
r2 = r2_score(serie_covid['Frecuencia'], fitted_values)
print("Coeficiente de determinacin (R):", r2)
```

Como se esperaba por los rdenes, la serie est sobre-ajustada

```python
forecast = resultado.get_forecast(steps=30)
pred_media = forecast.predicted_mean
conf_int = forecast.conf_int()

plt.figure(figsize=(12,5))
plt.plot(serie_covid['Frecuencia'], label='Serie observada')
plt.plot(pred_media.index, pred_media, label='Pronstico 30 das', color='red')
plt.fill_between(conf_int.index,
                 conf_int.iloc[:,0],
                 conf_int.iloc[:,1], color='pink', alpha=0.3)

plt.title('Pronstico con modelo SARIMA estacional (periodo = 237 das)')
plt.xlabel('Fecha')
plt.ylabel('Frecuencia')

# --- Dibujar lneas verticales cada 237 das ---
inicio = serie_covid.index.min()
fin = serie_covid.index.max()
```

Un ajuste perfecto de la serie no implica que las predicciones sean adeacudas. Se observa que hay das donde la media de las predicciones son negativas, lo cual biolgicamente no tiene sentido.

La leccin es que:
1.   No debemos 'casarnos' con un modelo, o con un tipo de modelo, o con una estructura de modelo. Cada conjunto de datos tiene modelos distintos que pueden ser adecuados tanto para el cumplimiento de supuestos, como para el ajuste y finalmente para la prediccin.
2.   Debe haber un balance entre: buen ajuste, predicciones razonables para el contexto de los datos y un modelo que est adecuadamente implementado.
