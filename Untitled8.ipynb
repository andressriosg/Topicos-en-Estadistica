```python
dummi_menores = pd.get_dummies(datos_menores1, columns = ["P51", "P52", "P55", "P56", "P57", "P6159s8", "P6163s8"], dtype = float)
dummi_menores.head(5) # P6161 son los datos de menores que no tienen crecimiento y desarrollo
```

```python
dummi_menores.columns
```

```python
dummi_menores.rename(columns = {'P51_2':'acu casa', 'P51_3':'acu trab', 'P51_4':'niñera', 'P51_5':"mayor", 'P51_6':"menor", 'P51_7':"solo", 'P51_8':"otra persona"}, inplace = True)
dummi_menores.rename(columns = {'P52_1.0':"lejos", 'P52_2.0':"costoso", 'P52_3.0':"sin cupo", 'P52_4.0':"casa", 'P52_5.0':"edad", 'P52_6.0':"otro motivo"}, inplace = True)
dummi_menores.rename(columns = {'P55_1':"si desa", 'P55_2':"no desa", 'P56_1':"si alm", 'P56_2':"no alm", 'P57_1':"si onces", 'P57_2':"no onces"}, inplace = True)
dummi_menores.rename(columns = {'P6159s8_1.0':"no comp madre", 'P6159s8_2.0':"si comp madre", 'P6163s8_1.0':"no comp padre", 'P6163s8_2.0':"si comp padre"}, inplace = True)
dummi_menores.head(5)
```

```python
dummi_menores.P6161 = dummi_menores["P6161"].astype(str) # P6161 corresponde a valores de una
#variable cualitiva
dummi_menores.P6161.loc[(dummi_menores.P6161 == "1")] = "Si"
dummi_menores.P6161.loc[(dummi_menores.P6161 == "2")] = "No"
dummi_menores.P6161.head(2)
```

```python
X = dummi_menores.drop(["P6161", 'P6161s1'], axis = 1) # Variables explicativas
y = dummi_menores.P6161 # Variable respuesta
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
y_train
```

Hemos escogido, por validación cruzada, el 20% de datos como validación, y por tanto, 80% de datos para el entrenamiento.

**Red neuronal para clasificación**

```python
from sklearn.neural_network import MLPClassifier
mlp_clf = MLPClassifier(hidden_layer_sizes=(50,50,50, 50),
                        max_iter = 500, activation = 'logistic',
                        solver = 'sgd') # Red neuronal de percentrón multicapa para clasificación
# solver = sgd es el metodo de gradiente descentende estocastico. solver = adam es un método derivado de sgd
```

Se implementó una red neuronal de perceptrón multicapa para clasificación con 4 capas ocultas, cada una de 50 nodos. Máximo hará 500 iteraciones, la función de activación es la logística. Se van a estimar los parámetros por gradiente descendente estocástico. Recomendación: usar solver = "sgd" o solver = "adam", para obtener la *curva de pérdida*.

```python
mlp_clf.fit(X_train, y_train) # fase de entrenamiento se hace sobre los datos de entrenamiento
```

```python
y_pred = mlp_clf.predict(X_test) # Fase de validación se hace con los datos de validación
from sklearn.metrics import accuracy_score
print('Precisión: {:.2f}'.format(accuracy_score(y_pred, y_test)))
```

83% de los valores se predicen acertdamente.

```python
from sklearn.metrics import confusion_matrix
import pylab as pl
import seaborn as sns
import matplotlib.pyplot as plt

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)

# Crear gráfico
plt.figure()
ax = plt.subplot()

sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', ax=ax)

# Etiquetas
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')

# Etiquetas de clases
ax.set_xticklabels(['No', 'Sí'])
ax.set_yticklabels(['No', 'Sí'])

plt.show()
```

94 menores que no reciben atención, de acuerdo con la predicción del modelo si la reciben. Una matriz de confusión adecuada es aquella cuyos elementos fuera de la diagonal son cero

**Clasificador gausiano Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
clase_bayes = GaussianNB(var_smoothing = 1e-30)
clase_bayes.fit(X_train, y_train)
y_pred_bayes = clase_bayes.predict(X_test)
print('Accuracy: {:.2f}'.format(accuracy_score(y_pred_bayes, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
import pylab as pl
cm2 = confusion_matrix(y_test, y_pred_bayes)
cm2
import seaborn as sns
import matplotlib.pyplot as plt
ax= plt.subplot()
sns.heatmap(cm2, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation
# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
ax.set_title('Confusion Matrix')
```

```python
from sklearn.metrics import precision_score
precision_por_clase = precision_score(y_test, y_pred_bayes, average=None)
precision_por_clase
```

**Árboles de decisión**

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred_tree = clf.predict(X_test)
print('Accuracy: {:.2f}'.format(accuracy_score(y_pred_tree, y_test)))
```

```python
cm3 = confusion_matrix(y_test, y_pred_tree)
cm3
ax= plt.subplot()
sns.heatmap(cm3, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation
# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
ax.set_title('Confusion Matrix')
```

**Análisis discriminante cuadrático**

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
clf_qua = QuadraticDiscriminantAnalysis()
clf_qua.fit(X_train, y_train)
y_pred_qua = clf_qua.predict(X_test)
print('Accuracy: {:.2f}'.format(accuracy_score(y_pred_qua, y_test)))
```

```python
cm4 = confusion_matrix(y_test, y_pred_qua)
ax= plt.subplot()
sns.heatmap(cm4, annot=True, fmt='g', ax=ax);
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
ax.set_title('Confusion Matrix')
```

**Clasificador de Gradiente Boosting**

```python
from sklearn.ensemble import GradientBoostingClassifier
clf_boot = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
                                      max_depth=1, random_state=0)
clf_boot.fit(X_train, y_train)
y_pred_boot = clf_boot.predict(X_test)
print('Accuracy: {:.2f}'.format(accuracy_score(y_pred_boot, y_test)))
```

```python
cm5 = confusion_matrix(y_test, y_pred_boot)
ax= plt.subplot()
sns.heatmap(cm5, annot=True, fmt='g', ax=ax);
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
ax.set_title('Confusion Matrix')
```
